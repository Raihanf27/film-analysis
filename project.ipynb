```python
#Untuk membuang kolom gross dan runtime. Dan membuang nilai yang kosong pada dataframe pada kolom lainnya
def drop_na_movies (df):
    df = df.drop("Gross", axis=1)
    df = df.drop("RunTime", axis = 1)
    df = df.dropna()
    return df
```

```python
#Memanggil class drop_na_movies
df = drop_na_movies(df)
```

### 3. Memperbaiki kolom tahun

```python
df
```

```python
df['YEAR']
```

```python
# Hapus tanda kurung
df['YEAR'] = df['YEAR'].str.replace('[()]', '', regex=True).str.strip()
```

```python
# Mengganti Varian Tanda '–' dengan Tanda '-':
df['YEAR'] = df['YEAR'].str.replace('–', '-', regex=True)
```

```python
# Menghilangkan Tanda '-' Ganda:
df['YEAR'] = df['YEAR'].str.replace('-+', '-', regex=True)
```

```python
# Tambahkan '-' di akhir untuk nilai yang hanya memiliki satu tahun
df['YEAR'] = df['YEAR'].apply(lambda x: x if '-' in x else f"{x}-")
```

```python
df[['Start_Year', 'End_Year']] = df['YEAR'].str.split('-', expand=True)
```

```python
df['End_Year'] = df['End_Year'].replace({'': None})
df
```

```python
df = df.drop(columns='YEAR')
```

### 6. Membersihkan kolom stars

```python
df
```

```python
def stars_clean(df):
    df["STARS"] = df["STARS"].str.strip()
    df["Directors"] = pd.Series()
    df["Actors"] = pd.Series()
    for index in df.index:
        stars = df.loc[index, "STARS"]
        stars = stars.replace("\n", "")
        stars = stars.replace("Directors:", "")
        stars = stars.replace("Director:", "")
        stars = stars.replace("Stars:", "")
        df.at[index, "STARS"] = stars
        if "|" in stars:
            stars = stars.split("|")
            directors = stars[0].strip()
            df.at[index, "Directors"] = directors
            stars = stars[1].strip()
        df.at[index, "Actors"] = stars
    return df
```

```python
df = stars_clean(df)
```
